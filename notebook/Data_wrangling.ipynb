# Data_wrangling

```python
import pandas as pd
import numpy as np
import re
import warnings
warnings.filterwarnings('ignore')

#Reading the datasets
personality_data = pd.read_csv('personality_scores.csv', sep=';')
department_data = pd.read_csv('departments.csv', sep=';')
```

```python
personality_data
```

### Examining the dataframe by ID and droppping duplicates 

```python
#checking the duplicates
personality_new= personality_data.drop_duplicates(subset = 'ID')
```

### Assert statement to check that the new data frame is the length of the unique entries of the original data frame.

```python
# Asserting new personality data frame with the original dataframe
personality_new['ID'].unique().sum() == personality_data['ID'].sum()
```

### Droppping columns with all null values

```python
personality_new=personality_new.dropna(axis = 1)
personality_new.head(3)
```

### Renaming columns by replacing the columns name by texts inside the '[]' of each column

```python
personality_new.rename(columns=lambda x: x.replace('Section 5 of 6 [', ''), inplace=True)
personality_new.rename(columns=lambda x: x.replace('.]',''), inplace=True)
personality_new.rename(columns=lambda x: x.replace(']',''), inplace=True)

personality_new.head(3)
```

## Creating new columns containing the total score of each of the personality test subscales.

```python
def score(row, number, dataframe):
    '''Returns a count with
    row = 'row interations
    number = 'Personality it belongs to'
    dataframe = "new dataframe"
    '''
    count = 0
    for column in dataframe.columns:
        breakdown = row[column]
        if type(breakdown) == str:
            tup_break = breakdown[1]
            result = int(breakdown[4])
            if tup_break == number:
                count += result
    return count

#Adding columns to the dataframe and applying the score function
personality_new['Extraversion'] = personality_new.apply(lambda row : score(row,'1', personality_new), axis = "columns")
personality_new['Agreeableness'] = personality_new.apply(lambda row : score(row,'2', personality_new), axis = "columns")
personality_new['Conscientiousness'] = personality_new.apply(lambda row : score(row,'3', personality_new), axis = "columns")
personality_new['Emotional stability'] = personality_new.apply(lambda row : score(row,'4', personality_new), axis = "columns")
personality_new['Intellect'] = personality_new.apply(lambda row : score(row,'5', personality_new), axis = "columns")

display(personality_new)
```

### Merging the personality_new with department_data

```python
merged_data = pd.merge(department_data, personality_new, how='left')
merged_data
```

# 2. FILTERING the merged data frame so that only the applicants who scored less than 30 on emotional stability, conscientiousness AND agreeableness remain. Printing the ID numbers and departments of these applicants to the screen, and also assigning these applicants the tag “high_risk” in a new column. All other applicants get the tag “low_risk”

#### Filtering the merged data

```python
filtered_high = merged_data[(merged_data.Conscientiousness < 30) & (merged_data.Agreeableness < 30) & (merged_data['Emotional stability'] < 30)]
filtered_high

```

#### Creating a new 'Risk' column and assigning "low_risk" and "high_risk" tags on merged_data

```python
merged_data['Risk']=np.where((merged_data.Conscientiousness < 30) & (merged_data.Agreeableness<30) & (merged_data['Emotional stability'] <30),'high_risk','low_risk')
merged_data.head()
```
